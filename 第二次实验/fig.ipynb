```python
import numpy as np

import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 梁横截面参数
height = 20  # 高度，单位mm
width = 5    # 宽度，单位mm

# 取高度方向上的点
y = np.linspace(-height/2, height/2, 100)
# 假设最大弯矩M=1，惯性矩I=1，弯曲正应力公式σ = My/I
M = 1
I = 1
sigma = M * y / I

plt.figure(figsize=(6, 4))
plt.plot(sigma, y, label='弯曲正应力分布')
plt.axvline(0, color='gray', linestyle='--')
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel('弯曲正应力 σ')
plt.ylabel('截面高度 y')
plt.title('梁横截面上弯曲正应力分布规律')
plt.legend()
plt.grid(True)
plt.show()
```
